```python
a,b=1,2
def function(x):
    return x**3-x**2-1

def f1(x):
    return 3*x**2-1

def derivative2(x):
    return 6*x
```

```python
def bitection(f,a,b,tol=1e-6,max=1000):
    """f is objective function whose root is determiend"""
    if f(a) * f(b) >= 0:
        raise ValueError("it must be opposite in sign")

    for i in range(max):
        c=(a+b) / 2
        if abs(f(c))<tol or (b-a)/2<tol:
            return c,i
        if f(c) * f(a) <0:
            b=c
        else:
            a=c
    return c,max

    
```

```python
root,iteration=bitection(function,a,b,1e-6,40)
print(f"root is {root:6f},{iteration}")
```

```python
def newton_raphson(fun,f,x0,tol=1e-6,max=1000):
    x=x0
    for i in range(max):
        dfx=f(x)
        if dfx==0:
            raise ZeroDivisionError("derivative is zero")

        x_ne=x- fun(x)/dfx
        if abs(x_ne-x)<tol:
            return x_ne,i
        x=x_ne
    return x,max
```

```python
x,itr=newton_raphson(function,f1,3,1e-6,40)
print(f"root is {x}")
```
