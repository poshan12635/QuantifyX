```python
import pandas as pd
import os
```

```python
file_pat='data'
all_files=os.listdir(file_pat)
```

```python
dfs=[]
for file in all_files:
    if file.endswith('.csv'):
        file_path=os.path.join(file_pat,file)
        df=pd.read_csv(file_path)
        dfs.append(df)
        
merge_df=pd.concat(dfs,ignore_index=True)   
    
```

```python
merge_df.isnull().sum()
```

```python
merge_df.dtypes
```

```python
merge_df['Volume']=(merge_df['Volume'].replace('-',None).str.replace(',','',regex=False).astype(float))
```

```python
merge_df.isnull().sum()
```

```python
merge_df['Date']=pd.to_datetime(merge_df['Date'])
```

```python
merge_df.columns
```

```python
df2 = merge_df.drop(columns=['Percent Change', 'Turn Over'])
```

```python
df2.isnull().sum()
df2=df2.sort_values(['Symbol','Date'])
```

```python
df2['Return']=df2.groupby('Symbol')['Close'].pct_change()
```

```python
df2
```

```python
df2['Return'].max()
```

```python
df2.to_csv('hydropower(1year).csv')
```
