```python
folder_path='data'
all_files=os.listdir(folder_path)
```

```python
dfs=[]
for file in all_files:
    if file.endswith(".csv"):
        file_path=os.path.join(folder_path,file)
        df=pd.read_csv(file_path)
        dfs.append(df)


merge_df=pd.concat(dfs,ignore_index=True)
merge_df.to_csv("merged.csv",index=False)
```

```python
df2=pd.read_csv('merged.csv')
```

```python
df2
```

```python
df2['Return']=df2['Close'].pct_change()
```

```python
df2
```

```python
df2.dropna()
```

```python
volatility=(df2.groupby('Symbol')['Return'].std().sort_values(ascending=False))
```

```python
print(volatility.head(10))
```

```python
vol_df1=volatility.reset_index()
vol_df1.columns=['Symbol','Volatility']
```

```python
vol_df1
```

```python
average_return=(df2.groupby('Symbol')['Return'].mean().sort_values(ascending=False))
```

```python
ret_df1=average_return.reset_index()
ret_df1.columns=['Symbol','Average_Return']
```

```python
ret_df1
```

```python
merged_df=pd.merge(vol_df1,ret_df1,on='Symbol')
```

```python
merged_df
```

```python
merged_df['sharpe_ratio_daily']=merged_df['Average_Return']/merged_df['Volatility']
```

```python
merged_df['sharpe_ratio_anuually']=merged_df['sharpe_ratio_daily'] *np.sqrt(252)
```

```python
merged_df.to_csv('hydro_return.csv',index=True)
```
